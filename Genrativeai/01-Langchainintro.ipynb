```python
import os
import sys
from dotenv import load_dotenv

# अगर नीचे वाला ImportError आए तो Kernel गलत है — Cursor में Kernel: "Python (Langchain .venv)" चुनें + Restart
print("Python:", sys.executable)
try:
    from langchain.chat_models import init_chat_model
except ImportError as e:
    raise ImportError(
        f"{e}\n\nChalane wala interpreter: {sys.executable}\n"
        "Kernel → Pick 'Python (Langchain .venv)' → Restart Kernel."
    ) from e

load_dotenv(r"C:\Langchain\.env")
print("API Key:", "OK" if os.getenv("GOOGLE_API_KEY") else "missing")

model = init_chat_model("gemini-flash-latest", model_provider="google_genai")
response = model.invoke("नमस्ते, आप कैसे हैं?")
content = response.content
if isinstance(content, list):
    content = "".join(
        part.get("text", "") if isinstance(part, dict) else str(part)
        for part in content
    )
print(content)
```

#### Genrative Ai Application with gemini model


```python
import importlib
import os
import sys
from dotenv import load_dotenv

_m = sys.modules.get("langchain_core.exceptions")
if _m is not None and not hasattr(_m, "ContextOverflowError"):
    importlib.reload(_m)

from langchain_google_genai import ChatGoogleGenerativeAI

print("Python:", sys.executable)

# अपनी .env फ़ाइल का सही पाथ यहाँ दें
env_path = r"C:\Langchain\.env"
load_dotenv(dotenv_path=env_path)
print("API Key Load Status:", "Success" if os.getenv("GOOGLE_API_KEY") else "Failed")

# gemini-flash-latest = चालू Flash (purana gemini-1.5-flash nayi API par aksar 404)
llm = ChatGoogleGenerativeAI(model="gemini-flash-latest")

response = llm.invoke("नमस्ते, कैसे हो?")
content = response.content
if isinstance(content, list):
    content = "".join(
        part.get("text", "") if isinstance(part, dict) else str(part)
        for part in content
    )
print(content)
```
